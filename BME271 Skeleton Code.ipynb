```python
import numpy as np
import cv2 as cv
import cv2 as cv2
from matplotlib import pyplot as plt

img = cv.imread('/Users/ashwingadiraju/Documents/BME271/Discussion 4/Eagle.jpeg',0)
#img = cv.imread('/Users/ashwingadiraju/Documents/BME271/Discussion 4/Elephant.png',0)
#img = cv.imread('/Users/ashwingadiraju/Documents/BME271/Discussion 4/Squirrel.jpeg',0)
#img = cv.imread('/Users/ashwingadiraju/Documents/BME271/Discussion 4/bird.jpeg',0)
edges = cv.Canny(img,100,400)

plt.imshow(img,cmap = 'gray')
plt.title('Original Image'), plt.xticks([]), plt.yticks([])
```

```python
img_blur = cv2.medianBlur(img, 31)
edges = cv.Canny(img_blur,1,100)

plt.imshow(255-edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img_blur = cv2.medianBlur(img, 11)
edges = cv.Canny(img_blur,1,100)

plt.imshow(255-edges,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])

plt.show()
```

```python
img_blur = cv2.medianBlur(img, 71)

img_blur2 = cv2.medianBlur(img, 11)
edges = cv.Canny(img_blur2,1,100)
edges = edges * 70
inv_img = 255-edges

plt.imshow(inv_img+img_blur,cmap = 'gray')
plt.title('Edge Image'), plt.xticks([]), plt.yticks([])
```
